```python
import glob
import fitz
from IPython.display import Image, display

pdf_files = sorted(glob.glob('../DesignSpaceExploration/*.pdf'))

for pdf_path in pdf_files:
    print(pdf_path)
    doc = fitz.open(pdf_path)
    page = doc[0]
    
    im = page.get_pixmap()
    display(Image(im.tobytes()))
```
