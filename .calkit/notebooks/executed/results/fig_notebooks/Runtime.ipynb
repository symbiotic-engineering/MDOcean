```python
import glob
import fitz
from IPython.display import Image, display

pdf_files = sorted(glob.glob('../Runtime/*.pdf'))

for pdf_path in pdf_files:
    print(pdf_path)
    doc = fitz.open(pdf_path)
    for page in doc:
        im = page.get_pixmap()
        display(Image(im.tobytes()))
    doc.close()
```
